## How to install Jupyter
---

### Instructions on Jupyter website [here](http://jupyter.readthedocs.org/en/latest/install.html#existing-python-new-jupyter)
### For Mac:
1. Open Terminal window
2. Run command: ```>>> pip install jupyter```      (if you are using Python 3.0 change `pip` to `pip3`)
3. Test install by typing in Terminal: ```>>> jupyter notebook```

### Troubleshooting:
1. **pip not installed:** *Solution* install pip
2. **unable to find jupyter install after pip install:** *Solution* find jupyter manually, or choose alternative install method for jupyter such as Anaconda
3. **unwillingness to install Anaconda:** *Solution* persistence

## Instructions for optional ways to install on Juila website [here](http://julialang.org)

## For Mac:
1. Download binary file for your version of OSX [here](http://julialang.org/downloads/).
2. Open .dmg file and drag Julia application to Applications folder.
3. In Terminal, try initiating a session of Julia by typing: ```>>> julia```

### Troubleshooting
* **Error `julia: command not found`** is likely a pathing issue.  To fix, tell your computer where bin/julia is by
    1. Adding a soft link to `path/to/install/location/julia/usr/bin` in location a folder already in your `PATH` environment variable
    2. Appending `path/to/install/location/julia/usr/bin` to your `PATH` variable.

## How to install iJulia (using Julia in Jupyter!)
---
### For Mac:
1. Install Jupyter
2. Install Julia
3. In Terminal start julia: `>>> julia`
4. Add the iJulia package by then running the Julia command: `Pkg.add("IJulia")`
5. To test in Terminal start a Jupyter session: `jupyter notebook`
6. Add new file in Jupyter using dropdown menu in upper right.
7. If you already have a notebook you want to open then: `jupyter notebook mynotebook.ipynb`

## Mandelbrot set for a (very) brief intro to Julia
### (using packages, functions, for-loops, benchmarking, and plotting)

````julia
# Look, we can import Python packages (Pkg.add("PyPlot"))
using PyPlot
````

````julia
# An example of a function from a tutorial by David Sanders on YouTube from SciPy 2014
#
# A julia set 
# (A mandelbrot set is obtained by setting z==c initially)
# This code is from the Julia tutorial on 
function julia(z,c; maxiter=200)
    for n=1:maxiter
        if abs2(z)>4
            return n-1
        end
        z=z*z+c
    end
    return maxiter
end
````

````julia
# here we time a for loop calling the julia set function
@time m = [ UInt8(julia(complex(r,i), complex(-.51,.711))) for i=1:-.002:-1.0, r=-1.5:.002:1.5];
````

````julia
# the notebook is able to display ColorMaps
get_cmap("RdGy")
````

````julia
# show an image
f=imshow(m,cmap="RdGy")
title("Title's are nice")
````
